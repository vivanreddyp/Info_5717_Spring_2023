(1) (10 points) Calculate the number of words that start with a vowel (a, e, i, o, and u). Print out a list of these words, along with their frequency (number of times they appear in the text). The list should be in decreasing order of frequency.
(2) (10 points) Print out words that have an apostrophe (‘) in the middle of the word. Examples include English contractions such as “didn’t”, “wasn’t”, “couldn’t", or "she'd". 
(3) (10 points) Remove all stopwords, and words with an apostrophe in the text and print it out. You can find a list of English stopwords here: http://www.ranks.nl/stopwords. 


```python
import re
from collections import Counter

text = "Harry lay in his dark cupboard much later, wishing he had a watch. He didn't know what time it was and he couldn't be sure the Dursleys were asleep yet. Until they were, he couldn't risk sneaking to the kitchen for some food. He'd lived with the Dursleys almost ten years, ten miserable years, as long as he could remember, ever since he'd been a baby and his parents had died in that car crash. He couldn't remember being in the car when his parents had died. Sometimes, when he strained his memory during long hours in his cupboard, he came up with a strange vision: a blinding flash of green light and a burning pain on his forehead. This, he supposed, was the crash, though he couldn't imagine where all the green light came from. He couldn't remember his parents at all. His aunt and uncle never spoke about them, and of course he was forbidden to ask questions. There were no photographs of them in the house."

# Find all words starting with vowels (case-insensitive)
vowels_regex = r'\b[aeiou]\w*'
vowels = re.findall(vowels_regex, text, flags=re.IGNORECASE)

# Count frequency of vowel words
vowel_freq = Counter(vowels)

# Print frequency of vowel words in descending order
print("Vowel words frequency:")
for word, count in sorted(vowel_freq.items(), key=lambda x: x[1], reverse=True):
    print(f"{word}: {count}")
```

(1)	(10 points). Create two files based on data published at U.S. climate data (http://www.usclimatedata.com/climate/denton/texas/united-states/ustx0353):

(2)	(20 points). Write a program to calculate the mean, median, and standard deviation of high temperature, low temperature and precipitation of each file, and output the results in the following format:
File name		   mean			median		standard deviation
2011-Jan-June.txt               -----			  -----			-----
2018-Jan-June.txt


```python
import pandas as pd
import numpy as np

# read in the data files
file_a = pd.read_csv('2011-Jan-June.csv')
file_b = pd.read_csv('2018-Jan-June.csv')

# define a function to calculate mean, median, and standard deviation for a given column
def calculate_stats(df, col):
    return {
        'mean': round(np.mean(df[col]), 2),
        'median': round(np.median(df[col]), 2),
        'std': round(np.std(df[col]), 2)
    }

# calculate statistics for each file and column
stats_a = {}
stats_a['High'] = calculate_stats(file_a, 'High')
stats_a['Low'] = calculate_stats(file_a, 'Low')
stats_a['Precip'] = calculate_stats(file_a, 'Precip')

stats_b = {}
stats_b['High'] = calculate_stats(file_b, 'High')
stats_b['Low'] = calculate_stats(file_b, 'Low')
stats_b['Precip'] = calculate_stats(file_b, 'Precip')

# print the results
print('File name\t\tmean\t\tmedian\t\tstandard deviation')
print('2011-Jan-June.csv\t{}\t{}\t{}'.format(stats_a['High']['mean'], stats_a['High']['median'], stats_a['High']['std']))
print('\t\t\t{}\t{}\t{}'.format(stats_a['Low']['mean'], stats_a['Low']['median'], stats_a['Low']['std']))
print('\t\t\t{}\t{}\t{}'.format(stats_a['Precip']['mean'], stats_a['Precip']['median'], stats_a['Precip']['std']))
print('2018-Jan-June.csv\t{}\t{}\t{}'.format(stats_b['High']['mean'], stats_b['High']['median'], stats_b['High']['std']))
print('\t\t\t{}\t{}\t{}'.format(stats_b['Low']['mean'], stats_b['Low']['median'], stats_b['Low']['std']))
print('\t\t\t{}\t{}\t{}'.format(stats_b['Precip']['mean'], stats_b['Precip']['median'], stats_b['Precip']['std']))
```

For example, if a student's name is Jennifer Lynn Meyers her euid could be "jlm8940".  Use this function to print out the names and euids of the students you created in (2).

(4)	(8 pts) Write a function for the Student class called "register". This function asks the user to enter a class number, a semester, and a grade for each student. Then this information is added to the Student object. Call this function for the students you created in part (2). Each student should be registered for at least two courses but can have more than.

(5)	(8 pts) Write a function for the Student class that calculates a student's GPA and prints it out. Use it to print out the names and GPAs of the student's you created in (2). 


```python
import random

class Student:
    def __init__(self, first_name, last_name, middle_name=''):
        self.first_name = first_name
        self.last_name = last_name
        self.middle_name = middle_name
        self.euid = ''
        self.gpa = 0
        self.classes_taken = []

    def generate_euid(self):
        self.euid = self.first_name[:3].lower() + self.middle_name[:3].lower() + self.last_name[:3].lower() + str(random.randint(1000, 9999))

    def register(self, class_num, semester, grade):
        self.classes_taken.append({'class_num': class_num, 'semester': semester, 'grade': grade})

    def calculate_gpa(self):
        total_credits = 0
        total_grade_points = 0
        for course in self.classes_taken:
            grade = course['grade'].upper()
            if grade == 'A':
                grade_points = 4.0
            elif grade == 'B':
                grade_points = 3.0
            elif grade == 'C':
                grade_points = 2.0
            elif grade == 'D':
                grade_points = 1.0
            else:
                grade_points = 0.0
            total_credits += 1
            total_grade_points += grade_points
        self.gpa = round(total_grade_points / total_credits, 2)
        

# Prompt the user for three students
students = []
for i in range(3):
    first_name = input("Enter first name of student {}: ".format(i+1))
    middle_name = input("Enter middle name of student {}: ".format(i+1))
    last_name = input("Enter last name of student {}: ".format(i+1))
    student = Student(first_name, last_name, middle_name)
    student.generate_euid()
    students.append(student)
    print("Student {} created with euid: {}".format(i+1, student.euid))

# Register each student in at least two classes
for student in students:
    for i in range(2):
        print("Registering classes for student: ", student.first_name)
        class_num = input("Enter class number: ")
        semester = input("Enter semester: ")
        grade = input("Enter grade: ")
        student.register(class_num, semester, grade)

# Calculate and print the GPAs of each student
for student in students:
    student.calculate_gpa()
    print("Student {} - Name: {} GPA: {}".format(student.euid, student.first_name + ' ' + student.last_name, student.gpa))
```

```python

import random

# Define the four math question functions

def addition_question():
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    problem = f"{num1} + {num2}"
    solution = num1 + num2
    return problem, solution

def subtraction_question():
    num1 = random.randint(1, 10)
    num2 = random.randint(1, num1)
    problem = f"{num1} - {num2}"
    solution = num1 - num2
    return problem, solution

def multiplication_question():
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    problem = f"{num1} * {num2}"
    solution = num1 * num2
    return problem, solution

def division_question():
    num2 = random.randint(1, 10)
    num1 = num2 * random.randint(1, 10)
    problem = f"{num1} / {num2}"
    solution = num1 // num2
    return problem, solution

# Define the main quiz function

def run_quiz():
    # Display quiz introduction
    print("Welcome to the Math Quiz!")
    print("You will be presented with four different math questions:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("Enter the number of the question you want to answer (or 5 to exit).")

    # Initialize variables to keep track of results
    total = 0
    correct = 0

    # Loop until user chooses to exit
    while True:
        # Ask user which question to answer
        try:
            choice = int(input("Enter your choice: "))
            if choice == 5:
                break
            elif choice in [1, 2, 3, 4]:
                # Generate the question and get the user's answer
                if choice == 1:
                    problem, solution = addition_question()
                elif choice == 2:
                    problem, solution = subtraction_question()
                elif choice == 3:
                    problem, solution = multiplication_question()
                else:
                    problem, solution = division_question()
                print(f"What is {problem}?")
                user_answer = int(input("Enter your answer: "))
                # Check the user's answer and update the results
                if user_answer == solution:
                    print("Correct!")
                    correct += 1
                else:
                    print(f"Incorrect. The correct answer is {solution}.")
                total += 1
            else:
                print("Invalid input. Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 5.")
    
    # Display quiz results
    print(f"You answered {total} questions with {correct} correct.")
    percentage = (correct / total) * 100 if total > 0 else 0
    print(f"Your score is {percentage:.1f}%. Thank you for playing!")

# Call the main quiz function
run_quiz()
```
